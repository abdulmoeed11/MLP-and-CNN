```python
class OcularDataset(Dataset):
    
    # TODO:
    # Define constructor for AnimalDataset class
    # HINT: You can pass processed data samples and their ground truth values as parameters 
    def __init__(self, data, labels):
        self.data = torch.FloatTensor(data.astype('float')).cuda()
        self.labels = torch.FloatTensor(labels.astype('float')).cuda()
        

    '''This function should return sample count in the dataset'''
    def __len__(self):
        return self.labels.shape[0]

    '''This function should return a single sample and its ground truth value from the dataset corresponding to index parameter '''
    def __getitem__(self, index):
      # img=np.array(Image.open(self.root_dir +str(index)+'.jpg').convert('L')).reshape(1,-1)
      return self.data[index,:], self.labels[index]

def get_dataset(root_dir):
    # TODO: 
    # Construct training, validation and test sets
    labels = pd.read_excel('labels.xlsx')
    labels = np.array(labels) # labels.to_numpy()

    images = []
    # Read dataset files
    for j in labels[:,0]:
      # try:
      image = (Image.open(root_dir+str(j)+'.jpg').convert('L'))
      img_arr = np.array(image)
      images.append(img_arr)
      # except FileNotFoundError:
        # pass
    

    # Normalize & flatten datasets
    data = np.array(images).reshape(3500,-1)
    data, labels = shuffle(data, labels)
    train_dataset = data[0:2500,:]/255
    valid_dataset = data[2500:3000, :]/255
    test_dataset  = data[3000:3500, :]/255

    # labels configuration 
    labels = np.array(labels[:,1])

    train_labels = labels[0:2500]
    valid_labels = labels[2500:3000]
    test_labels = labels[3000:3500]
    
    # return train_dataset, val_dataset, test_dataset#
    return train_dataset, valid_dataset, test_dataset, train_labels, valid_labels, test_labels
```

```python
import torch.nn as nn
from sklearn.preprocessing import MinMaxScaler
import torch.nn.functional as F


class FNet(Module):
    '''Define your neural network'''
    def __init__(self, inputsize, outputsize): 
    # you can add any additional parameters you want 
    # TODO:
    # You should create your neural network here
      super(FNet, self).__init__()
      self.h1 = 1024
      self.h2 = 256
      self.inp = inputsize
      self.out = outputsize
      self.hidden_layer1 = nn.Linear(self.inp, self.h1)
      self.hidden_layer2 = nn.Linear(self.h1, self.h2)
      self.output = nn.Linear(self.h2, self.out)

      # self.optimizer = torch.optim.Adam(self.parameters(), lr=env.ACTION_SPACE)
      # self.lossFunc = nn.MSELoss()
      # self.device = torch.device('gpu')
      # self.to(self.device)
  
          
    def forward(self, x):
      
        # you can add any additional parameters you want
        # TODO:
        # Forward propagation implementation should be here
        # scaler = MinMaxScaler(feature_range=(0,1))
        # x = scaler.fit_transform(x)
        # x = max(0.0, x)
        # x = F.relu(x)
        x = F.relu(self.hidden_layer1(x))
        x = F.relu(self.hidden_layer2(x))
        return torch.sigmoid(self.output(x))
      
```

```python
import numpy as np
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
from torch.nn import BCELoss
from torch.nn import Module
from torchvision import transforms, utils
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
import torch.tensor as tens
import os
import torch.nn as nn
from sklearn.preprocessing import MinMaxScaler
import torch.nn.functional as F
from  sklearn.utils import shuffle

class OcularDataset_cnn(Dataset):
    # TODO:
    # Define constructor for SVHNDataset class
    # HINT: You can pass processed data samples and their ground truth values as parameters 
    def __init__(self, data, labels):
        self.data = torch.FloatTensor(data.astype('float')).cuda()
        self.labels = torch.FloatTensor(labels.astype('float')).cuda()


    '''This function should return sample count in the dataset'''
    def __len__(self):
       return self.data.shape[0]

    '''This function should return a single sample and its ground truth value from the dataset corresponding to index parameter '''
    def __getitem__(self, index):
      # img=np.array(Image.open(self.root_dir +str(index)+'.jpg').convert('L')).reshape(1,-1)
      return self.data[index,:,:], self.labels[index]
        

def get_dataset_cnn(root_dir):
    # TODO: 
    # Construct training, validation and test sets
    labels = pd.read_excel('labels.xlsx')
    labels = np.array(labels) # labels.to_numpy()

    images = []
    # Read dataset files
    for j in range(labels[-1,0]+1):
      try:
        image = (Image.open(root_dir+str(j)+'.jpg'))
        img_arr = np.array(image)
        images.append(img_arr)
      except FileNotFoundError:
        pass
    data = np.array(images)
    data = data.reshape(3500,3,256,128)
    data ,labels = shuffle(data,labels)
    # Normalize & flatten datasets
    train_dataset = data[0:2500,:,:,:]/255
    valid_dataset = data[2500:3000, :,:,:]/255
    test_dataset  = data[3000:3500, :,:,:]/255

    # labels configuration 
    labels = np.array(labels[:,1])
    labels = labels.reshape(3500,1)
    train_labels = labels[0:2500]
    valid_labels = labels[2500:3000]
    test_labels = labels[3000:3500]
    
    # return train_dataset, val_dataset, test_dataset#
    return train_dataset, valid_dataset, test_dataset, train_labels, valid_labels, test_labels
```

#### Convolutional Neural Network [8 pts]

Now implement your CNN. ConvNet class will represent your convolutional neural network. Implement 3 layers of convolution: 
1. <i>> 8 filters with size of 3 x 3 x 3 with stride 1 and no padding,</i><br> 
    <i>> ReLU </i><br>
2. <i>> 16 filters with size of 3 x 3 x 3 with stride 1 and no padding,</i><br>
    <i>> ReLU </i><br>
    <i>> MaxPool 2 x 2 </i><br>   
3. <i>> 32 filters with size of 3 x 3 x 3 with stride 1 and no padding,</i><br>
    <i>> ReLU </i><br>
    <i>> MaxPool 2 x 2 </i><br>

As a classification layer, you need to add only one more fully-connected layer at the end of the network. You need to choose the appropriate input and output neuron sizes and the activation function.

```python
class ConvNet(nn.Module):
    '''Define your neural network'''
    def __init__(self): # you can add any additional parameters you want 
    # TODO:
    # You should create your neural network here
      super(ConvNet, self).__init__()
      self.conv1 = nn.Conv2d(in_channels=3,out_channels=8,stride=1,padding=0,kernel_size=3)
      self.conv2 = nn.Conv2d(in_channels=8,out_channels=16,stride=1,padding=0,kernel_size=3)
      self.conv3 = nn.Conv2d(in_channels=16,out_channels=32,stride=1,padding=0,kernel_size=3)
      self.mp = nn.MaxPool2d(2)
      self.fc = nn.Linear(59520, 1)

    def forward(self, x): # you can add any additional parameters you want
    # TODO:
    # Forward propagation implementation should be here
      in_size = x.size(0)
      x = F.relu(self.conv1(x))
      x = F.relu(self.mp(self.conv2(x)))
      x = F.relu(self.mp(self.conv3(x)))
      x = x.view(-1,59520)  # flatten the tensor
      x = self.fc(x)
      # return F.log_softmax(x)
      return torch.sigmoid(x)
```
